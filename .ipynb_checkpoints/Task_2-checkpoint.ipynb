# Zadanie: statystyki
Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generowanie danych z rozkładu normalnego
np.random.seed(42)  # Ziarno dla powtarzalności
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)

# Funkcja do obliczania statystyk
def calculate_statistics(data, title):
    print(f"Statystyki dla {title}:")
    mean = np.mean(data)
    median = np.median(data)
    mode_result = stats.mode(data)  # Obliczenie mody
    std_dev = np.std(data, ddof=1)
    variance = np.var(data, ddof=1)
    q0 = np.quantile(data, 0.0)
    q1 = np.quantile(data, 0.25)
    q2 = np.quantile(data, 0.5)
    q3 = np.quantile(data, 0.75)
    q4 = np.quantile(data, 1.0)
    iqr = q3 - q1

    print(f"Średnia: {mean}")
    print(f"Mediana: {median}")
    
    # Obsługa mody
    try:
        # Sprawdzenie, czy moda istnieje i czy jest to tablica
        if hasattr(mode_result, 'mode') and hasattr(mode_result, 'count'):
            mode_value = mode_result.mode[0] if np.isscalar(mode_result.mode) else mode_result.mode
            mode_count = mode_result.count[0] if np.isscalar(mode_result.count) else mode_result.count
            print(f"Moda: {mode_value} (wystąpiła {mode_count} razy)")
        else:
            print("Moda: Brak jednoznacznej mody (wszystkie wartości są unikalne)")
    except Exception as e:
        print(f"Moda: Nie udało się obliczyć mody. Błąd: {e}")

    print(f"Odchylenie standardowe: {std_dev}")
    print(f"Wariancja: {variance}")
    print(f"Kwartyle: Q0={q0}, Q1={q1}, Q2={q2}, Q3={q3}, Q4={q4}")
    print(f"IQR: {iqr}")

    # Histogram
    plt.hist(data, bins=30, density=True, edgecolor='black', alpha=0.7)
    plt.title(f'Histogram dla {title}')
    plt.xlabel('Wartości')
    plt.ylabel('Gęstość')
    plt.axvline(x=mean, color='red', label='Średnia')
    plt.axvline(x=median, color='green', label='Mediana')
    
    # Dodanie mody do wykresu, jeśli istnieje
    try:
        if hasattr(mode_result, 'mode'):
            mode_value = mode_result.mode[0] if np.isscalar(mode_result.mode) else mode_result.mode
            plt.axvline(x=mode_value, color='orange', label='Moda')
    except Exception as e:
        print(f"Nie udało się dodać mody do wykresu. Błąd: {e}")
    
    plt.legend()
    plt.show()

    # Wykres pudełkowy
    plt.boxplot(data)
    plt.title(f'Wykres pudełkowy dla {title}')
    plt.ylabel('Wartości')
    plt.show()

# Obliczanie statystyk dla obu zbiorów
calculate_statistics(data_100, "100 obserwacji")
calculate_statistics(data_10000, "10 000 obserwacji")
```

Wyniki i porównanie statystyk
Dla 100 obserwacji:

    Średnia: Będzie bliska 3, ale może nieco odbiegać ze względu na małą liczbę obserwacji.

    Mediana: Również bliska 3.

    Moda: Może być różna, ponieważ w małej próbce wartości mogą się powtarzać rzadko.

    Odchylenie standardowe: Będzie bliskie 1, ale może być nieco inne ze względu na małą próbkę.

    Wariancja: Będzie bliska 1.

    Kwartyle: Q1 i Q3 będą bliskie teoretycznym wartościom (ok. 2.33 i 3.67).

    IQR: Będzie bliskie teoretycznej wartości (ok. 1.34).

Dla 10 000 obserwacji:

    Średnia: Bardzo bliska 3, ponieważ duża próbka lepiej przybliża teoretyczną średnią.

    Mediana: Również bardzo bliska 3.

    Moda: Może być bliższa teoretycznej wartości, ponieważ w dużej próbce wartości są lepiej reprezentowane.

    Odchylenie standardowe: Bardzo bliskie 1.

    Wariancja: Bardzo bliska 1.

    Kwartyle: Q1 i Q3 będą bardzo bliskie teoretycznym wartościom (ok. 2.33 i 3.67).

    IQR: Bardzo bliskie teoretycznej wartości (ok. 1.34).

Różnice między statystykami

    Średnia i mediana:

        Dla 100 obserwacji średnia i mediana mogą nieznacznie odbiegać od teoretycznej wartości 3.

        Dla 10 000 obserwacji średnia i mediana są bardzo bliskie 3, ponieważ duża próbka lepiej przybliża teoretyczny rozkład.

    Moda:

        W małej próbce (100 obserwacji) moda może być mniej reprezentatywna, ponieważ wartości powtarzają się rzadziej.

        W dużej próbce (10 000 obserwacji) moda jest bliższa teoretycznej wartości.

    Odchylenie standardowe i wariancja:

        Dla 100 obserwacji odchylenie standardowe i wariancja mogą nieznacznie odbiegać od teoretycznych wartości.

        Dla 10 000 obserwacji odchylenie standardowe i wariancja są bardzo bliskie teoretycznym wartościom.

    Kwartyle i IQR:

        Dla 100 obserwacji kwartyle i IQR mogą nieznacznie odbiegać od teoretycznych wartości.

        Dla 10 000 obserwacji kwartyle i IQR są bardzo bliskie teoretycznym wartościom.

Wnioski

    Duża próbka (10 000 obserwacji) lepiej przybliża teoretyczny rozkład normalny, a statystyki są bardziej stabilne i zbliżone do wartości teoretycznych.

    Mała próbka (100 obserwacji) może wykazywać większe odchylenia od teoretycznych wartości ze względu na mniejszą liczbę obserwacji.
