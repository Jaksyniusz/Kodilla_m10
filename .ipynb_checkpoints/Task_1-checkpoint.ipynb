# Zadanie: analiza statystyczna
Dokonaj statystycznej analizy (obliczenie średniej, kwartyli czy wykresów pudełkowych) dla zmiennej Fare (opłata).

```python
# Manipulacja danymi
import numpy as np
import pandas as pd

# Wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# Statystyczna analiza danych
from scipy import stats

# Ściągnięcie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv('titanic.csv')
dataset

# Przygotowanie danych
fare_values = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
print(fare_values.shape)

# Histogram dla Fare
plt.hist(fare_values, bins=20)
plt.title('Histogram dla Fare')
plt.xlabel('Fare')
plt.ylabel('Częstość')
plt.show()

# Średnia dla Fare
mean_fare = np.round(np.mean(fare_values))
print(f'Średnia Fare: {mean_fare}')

# Histogram z zaznaczoną średnią
plt.hist(fare_values, bins=20)
plt.axvline(x=mean_fare, color='red', label='Średnia')
plt.title('Histogram dla Fare z zaznaczoną średnią')
plt.xlabel('Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()

# Mediana dla Fare
median_fare = np.median(fare_values)
print(f'Mediana Fare: {median_fare}')

# Histogram z zaznaczoną średnią i medianą
plt.hist(fare_values, bins=20)
plt.axvline(x=mean_fare, color='red', label='Średnia')
plt.axvline(x=median_fare, color='green', label='Mediana')
plt.title('Histogram dla Fare z zaznaczoną średnią i medianą')
plt.xlabel('Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()

# Moda dla Fare
mode_fare = stats.mode(fare_values)
print(f'Moda Fare: {mode_fare}')

# Histogram z zaznaczoną średnią, medianą i modą
plt.hist(fare_values, bins=20)
plt.axvline(x=mean_fare, color='red', label='Średnia')
plt.axvline(x=median_fare, color='green', label='Mediana')
plt.axvline(x=mode_fare[0], color='orange', label='Moda')
plt.title('Histogram dla Fare z zaznaczoną średnią, medianą i modą')
plt.xlabel('Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()

# Kwartyle dla Fare
q0_fare = np.quantile(fare_values, 0.0)
q1_fare = np.round(np.quantile(fare_values, 0.25))
q3_fare = np.round(np.quantile(fare_values, 0.75))
q4_fare = np.round(np.quantile(fare_values, 1.0))
print(f'Q0: {q0_fare}')
print(f'Q1: {q1_fare}')
print(f'Q2: {median_fare}')
print(f'Q3: {q3_fare}')
print(f'Q4: {q4_fare}')

# Histogram z zaznaczonymi kwartylami
plt.hist(fare_values, bins=20)
plt.axvline(x=mean_fare, color='red', label='Średnia')
plt.axvline(x=median_fare, color='green', label='Mediana')
plt.axvline(x=mode_fare[0], color='orange', label='Moda')
plt.axvline(x=q0_fare, color='black', label='Min')
plt.axvline(x=q1_fare, color='black', label='Q1')
plt.axvline(x=q3_fare, color='black', label='Q3')
plt.axvline(x=q4_fare, color='black', label='Max')
plt.title('Histogram dla Fare z zaznaczonymi kwartylami')
plt.xlabel('Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()

# Zakres dla Fare
range_fare = max(fare_values) - min(fare_values)
print(f'Zakres Fare: {range_fare}')

# IQR dla Fare
iqr_fare = q3_fare - q1_fare
print(f'IQR Fare: {iqr_fare}')

# Wykres pudełkowy dla Fare
plt.boxplot(fare_values)
plt.title('Wykres pudełkowy dla Fare')
plt.ylabel('Fare')
plt.show()

# Wariancja dla Fare
variance_fare = np.var(fare_values, ddof=1)
print(f'Wariancja Fare: {variance_fare}')

# Odchylenie standardowe dla Fare
standard_deviation_fare = np.std(fare_values, ddof=1)
print(f'Odchylenie standardowe Fare: {standard_deviation_fare}')

# Standaryzacja Fare
scaler_standardized = StandardScaler()
standardized_fare_values = scaler_standardized.fit_transform(fare_values.reshape(-1, 1))
plt.hist(standardized_fare_values, bins=20)
plt.axvline(x=standardized_fare_values.mean(), color='red', label='Średnia')
plt.title('Histogram dla standaryzowanych wartości Fare')
plt.xlabel('Standaryzowane Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()
print(f'Średnia standaryzowanych wartości Fare: {standardized_fare_values.mean()}')
print(f'Odchylenie standardowe standaryzowanych wartości Fare: {np.std(standardized_fare_values, ddof=1)}')

# Normalizacja Fare
scaler_MinMax = MinMaxScaler()
normalized_fare_values = scaler_MinMax.fit_transform(fare_values.reshape(-1, 1))
plt.hist(normalized_fare_values, bins=20)
plt.axvline(x=normalized_fare_values.mean(), color='red', label='Średnia')
plt.title('Histogram dla znormalizowanych wartości Fare')
plt.xlabel('Znormalizowane Fare')
plt.ylabel('Częstość')
plt.legend(loc='upper right')
plt.show()
print(f'Średnia znormalizowanych wartości Fare: {normalized_fare_values.mean()}')
print(f'Odchylenie standardowe znormalizowanych wartości Fare: {np.std(normalized_fare_values, ddof=1)}')
```
